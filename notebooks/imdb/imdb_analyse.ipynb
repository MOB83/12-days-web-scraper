```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

df = pd.read_csv('../data/imdb/imdb_movies_clean.csv')

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,4))
ax1, ax2, ax3 = fig.axes

ax1.hist(df['imdb'], bins=10, range=(0,10))
ax1.set_title('IMDB Rating')

ax2.hist(df['metascore'], bins=10, range=(0,100))
ax2.set_title('Metascore')

ax3.hist(df['n_imdb'], bins=10, range=(0,100), histtype='step')
ax3.hist(df['metascore'], bins=10, range=(0,100), histtype='step')
ax3.legend(loc='upper left')
ax3.set_title('Normalized Distributions')

for ax in fig.axes:
	ax.spines['top'].set_visible(False)
	ax.spines['right'].set_visible(False)

plt.show()
```
